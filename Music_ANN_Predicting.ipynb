```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Input, LayerNormalization
from tensorflow.keras.losses import binary_crossentropy, categorical_crossentropy, logcosh, mean_squared_error, poisson, mean_absolute_error
from tensorflow.keras.optimizers import Adadelta, Adam, Nadam, RMSprop, SGD, Ftrl
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
pd.set_option('display.max_columns', None)
```

#### Load the dataset

```python
data = pd.read_excel('music data.xlsx')
```

```python
data.head()
```

```python
data.columns
```

```python
df = data[['danceability', 'energy',
       'speechiness','loudness', 'acousticness',
       'instrumentalness', 'liveness','tempo', 'duration_ms', 'valence','Popularity',
       'Versatility']]
```

```python
scaler = MinMaxScaler()
df[['danceability', 'energy',
       'speechiness','loudness', 'acousticness',
       'instrumentalness', 'liveness','tempo', 'duration_ms', 'valence','Popularity',
       'Versatility']] = scaler.fit_transform(df[['danceability', 'energy',
       'speechiness','loudness', 'acousticness',
       'instrumentalness', 'liveness','tempo', 'duration_ms', 'valence','Popularity',
       'Versatility']])
```

```python
df
```

```python
df = df.round(decimals = 1)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr().Versatility
```

```python
plt.scatter(x = df.Versatility, y = df.Popularity)
plt.show()
```

```python
versatility = df.Versatility
df = df.drop(columns = 'Versatility')
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, versatility, test_size = 0.2)
```

```python
x_train
```

```python
inputs = keras.Input(shape=(df.shape[1],))
x = layers.Dense(units = 512, activation = 'relu')(inputs)
x = layers.Dense(units = 256, activation = 'relu')(x)
x = layers.Dense(units = 128, activation = 'relu')(x)
x = layers.Dense(units = 64, activation = 'relu')(x)
outputs = layers.Dense(units = 1)(x)
model = keras.Model(inputs = inputs, outputs = outputs)
```

```python
opt = Adam(learning_rate = 0.001)
model.compile(loss = mean_absolute_error, optimizer = opt, metrics = ['mae'])
```

```python
fitting = model.fit(x = x_train, y = y_train, epochs = 2000, batch_size = 5)
```

```python
loss = fitting.history['loss']
epochs = range(len(loss))

plt.plot(loss)
plt.show()
```

```python
predictions = model.predict(x = x_test.values)
```

```python
score = model.evaluate(x=x_test, y = y_test)
score
```

#### Overfitted, let's stop learning after 400 epochs

```python
inputs = keras.Input(shape=(df.shape[1],))
x = layers.Dropout(rate = 0.03)(inputs)
x = layers.Dense(units = 128, activation = 'relu')(x)
x = layers.Dense(units = 128, activation = 'relu')(x)
x = layers.Dense(units = 128, activation = 'relu')(x)
x = layers.Dense(units = 128, activation = 'relu')(x)
outputs = layers.Dense(units = 1, activation = 'linear')(x)
model = keras.Model(inputs = inputs, outputs = outputs)
```

```python
opt = Adam(learning_rate = 0.001)
model.compile(loss = mean_absolute_error, optimizer = opt, metrics = ['mae'])
```

```python
fitting = model.fit(x = x_train, y = y_train, epochs = 400, batch_size = 5)
```

```python
loss = fitting.history['loss']
epochs = range(len(loss))

plt.plot(loss)
plt.show()
```

```python
predictions = model.predict(x = x_test.values)
```

```python
score = model.evaluate(x=x_test, y = y_test)
score
```
